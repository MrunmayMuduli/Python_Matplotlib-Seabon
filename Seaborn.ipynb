```python
sns.countplot(x="species", data=iris)
plt.show()
```

```python
n_setosa = iris[iris['species'] == 'setosa'].shape[0] 
n_versicolor = iris[iris['species'] == 'versicolor'].shape[0]
n_virginica = iris[iris['species'] == 'virginica'].shape[0]

data = [n_setosa, n_versicolor, n_virginica]
labels=['setosa','versicolor','virginica']

plt.pie(data,
       labels=labels,
       autopct='%.0f%%') # To show the portions in %ages

plt.show()
```

### **Summarizing Univariate analysis**
- Case 1: Numerical variable
  - Type of plots
    - Histogram
    - KDE
    - Box plot
- Case 2: Categorical variable
  - Type of plots
    - Barchart/Count plot
    - Pie chart

### Bivariate Plotting
1. NN
2. NC
3. CC

```python
# nn
sns.scatterplot(x=iris["petal_length"], y=iris["petal_width"])
plt.show()
#sns.scatterplot(x="petal_length", y="petal_width", data=iris)
```

```python
sns.lineplot(x=iris["petal_length"], y=iris["petal_width"])
plt.show()
```

```python
sns.jointplot(x="petal_length", y="petal_width", data=iris, kind="reg")
```

```python
sns.pairplot(data=iris)
```

```python
iris.corr()
```

```python
sns.heatmap(data=iris.corr(), annot=True)
```

```python
sns.heatmap(data=iris.corr(), annot=True, cmap="Blues")
```

```python
# case1: NN ---> Scatter Plot, Line Plot, Joint Plot
# Multiple NN pairs ---> Pairplot, corr --> heatmap
```

```python
# case2: NC
```

```python
sns.boxplot(x="species", y="sepal_length", data=iris)
plt.show()
```

```python
sns.kdeplot(x="sepal_length", data=iris, hue="species")
```

```python
sns.barplot(x="species", y="sepal_length", data=iris, estimator=np.mean)
```

```python
sns.violinplot(x="species", y="sepal_length", data=iris)
plt.show()
```
